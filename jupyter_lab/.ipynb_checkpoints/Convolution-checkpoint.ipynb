```python
import torch
from torch import nn
from torch.nn import functional as F
#Conv2d输入和输出格式固定为4D张量 batchsize channels height weight
conv2d = nn.Conv2d(1, 1, kernel_size = 3, padding = 1, stride = 2) #一般对称
net = nn.Sequential(conv2d)
X = torch.rand(8,8)
X = X.reshape((1, 1) + X.shape)
Y = net(X)
Y.shape
```
