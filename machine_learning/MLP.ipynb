```python
#给定输入 x, 权重 w, 偏移 b
#感知机输出: output = Activation Functions:σ(<w, x> + b)
'''
收敛定理 :
    · 数据在半径r内
    · 余量ρ分两类
        y(<w, x> + b) >= ρ

    ·保证在(r^2 + 1) / ρ^2 步内收敛
'''
#XOR问题(异或) 感知机无法拟合 --> 多层感知机
# —— —— —— —— —— —— —— —— —— —— —— —— —— —— —— —— —— ——
#输入层 -> 多隐藏层 -> 输出层 
#       超参数: ·隐藏层数 
#              ·每层隐藏层的大小
# 激活函数:防止隐藏层塌陷 


```

```python
import torch
from torch import nn
from d2l import torch as d2l

batch_size = 256
lr, num_epochs = 0.1, 10
train_iter, test_iter = d2l.load_data_fashion_mnist(batch_size)
```

```python
net = nn.Sequential(nn.Flatten(), nn.Linear(784, 256), nn.ReLU(),
                   nn.Linear(256, 10))

def init_weight(m) :
    if type(m) == nn.Linear :
        nn.init.normal_(m.weight, std = 0.1)
net.apply(init_weight)
```

```python
loss = nn.CrossEntropyLoss()
trainer = torch.optim.SGD(net.parameters(), lr = lr)
# d2l.train_ch3(net, train_iter, test_iter, loss, num_epochs, trainer) train_ch3 not in d2l??
```
